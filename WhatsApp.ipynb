```python
def enviar_whatsap():
    def enviar_msg():
        arquivo = open('linkchrome.txt', 'r')
        linhas = arquivo.readlines()
        options = webdriver.ChromeOptions()
        options.add_argument(r'user-data-dir={}'.format(linhas[0]))
        navegador_enviar = webdriver.Chrome(options=options)
        try:
            navegador_enviar.get(f"https://web.whatsapp.com/send?phone=+55{entry_telefone.get()}&text=Essa é uma menssagem automatica de envio de planilha (Segue sua planilha abaixo)")
            while True: 
                if len(navegador_enviar.find_elements_by_id('side')) < 1:
                    time.sleep(1)
                else:
                    time.sleep(2)
                    navegador_enviar.find_element_by_xpath('//*[@id="main"]/footer/div[1]/div/span[2]/div/div[2]/div[1]/div/div/p').send_keys(Keys.ENTER)
                    navegador_enviar.find_element_by_css_selector("span[data-icon='clip']").click()
                    attach = navegador_enviar.find_element_by_css_selector("input[type='file']")
                    attach.send_keys(var_caminho_arquivo_wp.get())
                    time.sleep(3)
                    send = navegador_enviar.find_element_by_css_selector("span[data-icon='send']")
                    send.click()
                    time.sleep(10)
                    break
        except:
            messagebox.showerror(title='Erro de envio de menssagem', message='Verifique a sua conexão com a internet, telefone ou arquivo selecionado.')
        arquivo.close()
    def selecionar_arquivo_whatsap():
        caminho_arquivo_wp = askopenfilename(title='Selecione a planilha')
        var_caminho_arquivo_wp.set(caminho_arquivo_wp)
        if button_selecionar_arquivo_whatsap:
            label_nome_arquivo_wp['text'] = f'Arquivo Selecionado: {caminho_arquivo_wp.split()[-1]}'
            label_nome_arquivo_wp['bg'] = 'yellow'
    def conectar_celular():
        arquivo = open('linkchrome.txt', 'r')
        linhas = arquivo.readlines()
        options = webdriver.ChromeOptions()
        options.add_argument(r'user-data-dir={}'.format(linhas[0]))
        navegador = webdriver.Chrome(options=options)
        navegador.get('https://web.whatsapp.com/')
        while True: 
            if len(navegador.find_elements_by_id('side')) < 1:
                time.sleep(1)
            else:
                label_telefone_conectado['text'] = 'CONECTADO'
                label_telefone_conectado['fg'] = '#43C22E'
        arquivo.close()

    janela_whatsap = tk.Toplevel()
    janela_whatsap.title('Janela Whatsapp')
    janela_whatsap.geometry("396x234")
    janela_whatsap.configure(bg = "#2a2a2a")
    canvas = Canvas(
        janela_whatsap,
        bg = "#2a2a2a",
        height = 234,
        width = 396,
        bd = 0,
        highlightthickness = 0,
        relief = "ridge")
    canvas.place(x = 0, y = 0)

    background_img = PhotoImage(file = f"bgwhatsapp.png")
    background = canvas.create_image(
        198.0, 97.5,
        image=background_img)

    img0 = PhotoImage(file = f"img0whatsapp.png")
    button_conectar_celular = Button(janela_whatsap,
        image = img0,
        borderwidth = 0,
        highlightthickness = 0,
        command = conectar_celular,
        relief = "flat")

    button_conectar_celular.place(
        x = 92, y = 42,
        width = 47,
        height = 26)

    img1 = PhotoImage(file = f"img1whatsapp.png")
    button_selecionar_arquivo_whatsap = Button(janela_whatsap,
        image = img1,
        borderwidth = 0,
        highlightthickness = 0,
        command = selecionar_arquivo_whatsap,
        relief = "flat")

    button_selecionar_arquivo_whatsap.place(
        x = 45, y = 150,
        width = 56,
        height = 33)

    entry0_img = PhotoImage(file = f"img_textBox0whatsapp.png")
    entry0_bg = canvas.create_image(
        207.5, 116.5,
        image = entry0_img)

    entry_telefone = Entry(janela_whatsap,
        bd = 0,
        bg = "#ffffff",
        highlightthickness = 0)

    entry_telefone.place(
        x = 82, y = 102,
        width = 251,
        height = 27)

    img2 = PhotoImage(file = f"img2whatsapp.png")
    button_enviar_msg = Button(janela_whatsap,
        image = img2,
        borderwidth = 0,
        highlightthickness = 0,
        command = enviar_msg,
        relief = "flat")

    button_enviar_msg.place(
        x = 152, y = 193,
        width = 92,
        height = 27)
    label_nome_arquivo_wp = tk.Label(janela_whatsap, text='NENHUM ARQUIVO SELECIONADO', font=("Arial", 8, "bold"), bg='#D9D9D9', fg='black', height=2, width=39)
    label_nome_arquivo_wp.place(x=100, y=150)
    label_telefone_conectado = tk.Label(janela_whatsap, text='VERIFICAR CONEXÃO', font=("Arial", 9, "bold"), bg='#2A2A2A', fg='white')
    label_telefone_conectado.place(x=156, y=46)

    janela_whatsap.resizable(False, False)
    janela_whatsap.mainloop()
```
